## Analysis Utilities: PCA & Power Spectral Density

Collect **24 samples per label** from the test set, then:
1. **PCA** — project channel-averaged PSD features to 2-D and scatter-plot Clear (0) vs Disruptive (1).
2. **PSD** — plot the average power spectral density for each label and the differential (Disruptive − Clear).

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy import signal

# ---------- config ----------
N_PER_LABEL = 24
FS = 500_000        # ECEi sampling rate in Hz (adjust if needed)
NPERSEG = 4096       # Welch segment length

# ---------- collect PSD features for each label ----------
psd_list_0, psd_list_1 = [], []

for i in range(len(dataset)):
    if len(psd_list_0) >= N_PER_LABEL and len(psd_list_1) >= N_PER_LABEL:
        break

    chunk, label = dataset[i]            # chunk: (20, 8, T)

    if label == 0 and len(psd_list_0) >= N_PER_LABEL:
        continue
    if label == 1 and len(psd_list_1) >= N_PER_LABEL:
        continue

    # flatten spatial dims → (160, T)
    flat = chunk.reshape(-1, chunk.shape[-1])

    # Welch PSD, then average across the 160 channels → (n_freqs,)
    freqs, psd = signal.welch(flat, fs=FS, nperseg=NPERSEG, axis=-1)
    psd_avg = psd.mean(axis=0)

    if label == 0:
        psd_list_0.append(psd_avg)
    else:
        psd_list_1.append(psd_avg)

psd_arr_0 = np.array(psd_list_0)   # (N0, n_freqs)
psd_arr_1 = np.array(psd_list_1)   # (N1, n_freqs)
print(f"Collected: {len(psd_arr_0)} clear (label 0), "
      f"{len(psd_arr_1)} disruptive (label 1)")
```

```python
# ── PCA on log-PSD features: Clear vs Disruptive ──────────────────────
all_psd = np.vstack([psd_arr_0, psd_arr_1])
all_labels = np.array([0] * len(psd_arr_0) + [1] * len(psd_arr_1))

# log-transform so PCA operates on order-of-magnitude differences
log_psd = np.log10(all_psd + 1e-30)

pca = PCA(n_components=2)
proj = pca.fit_transform(log_psd)

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(proj[all_labels == 0, 0], proj[all_labels == 0, 1],
           c='steelblue', s=80, edgecolors='k', linewidths=0.5,
           alpha=0.8, label='Clear (label 0)')
ax.scatter(proj[all_labels == 1, 0], proj[all_labels == 1, 1],
           c='firebrick', s=80, edgecolors='k', linewidths=0.5,
           alpha=0.8, label='Disruptive (label 1)')

ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
ax.set_title('PCA of Channel-Averaged log-PSD: Clear vs Disruptive')
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ── Average PSD per label + Differential PSD ──────────────────────────
psd_0_mean = psd_arr_0.mean(axis=0)
psd_1_mean = psd_arr_1.mean(axis=0)
psd_0_std  = psd_arr_0.std(axis=0)
psd_1_std  = psd_arr_1.std(axis=0)

freq_khz = freqs / 1e3

fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# ---- Clear (label 0) ----
axes[0].semilogy(freq_khz, psd_0_mean, color='steelblue', linewidth=1.5, label='Mean')
axes[0].fill_between(
    freq_khz,
    np.maximum(psd_0_mean - psd_0_std, 1e-30),
    psd_0_mean + psd_0_std,
    color='steelblue', alpha=0.2, label='± 1 std')
axes[0].set_title(f'Average PSD — Clear (n={len(psd_arr_0)})')
axes[0].set_xlabel('Frequency (kHz)')
axes[0].set_ylabel('PSD')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# ---- Disruptive (label 1) ----
axes[1].semilogy(freq_khz, psd_1_mean, color='firebrick', linewidth=1.5, label='Mean')
axes[1].fill_between(
    freq_khz,
    np.maximum(psd_1_mean - psd_1_std, 1e-30),
    psd_1_mean + psd_1_std,
    color='firebrick', alpha=0.2, label='± 1 std')
axes[1].set_title(f'Average PSD — Disruptive (n={len(psd_arr_1)})')
axes[1].set_xlabel('Frequency (kHz)')
axes[1].set_ylabel('PSD')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# ---- Differential PSD ----
diff_psd = psd_1_mean - psd_0_mean
axes[2].plot(freq_khz, diff_psd, color='black', linewidth=1.5)
axes[2].axhline(0, color='gray', linestyle='--', alpha=0.5)
axes[2].fill_between(
    freq_khz, 0, diff_psd,
    where=(diff_psd > 0), color='firebrick', alpha=0.3,
    label='Disruptive > Clear')
axes[2].fill_between(
    freq_khz, 0, diff_psd,
    where=(diff_psd < 0), color='steelblue', alpha=0.3,
    label='Clear > Disruptive')
axes[2].set_title('Differential PSD (Disruptive − Clear)')
axes[2].set_xlabel('Frequency (kHz)')
axes[2].set_ylabel('ΔPSD')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
